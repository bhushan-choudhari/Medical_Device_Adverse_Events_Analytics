```python
zip_file = '../data/device-event-0004-of-0004.json.zip'
with zipfile.ZipFile(zip_file, "r") as z:
    for filename in z.namelist():
        with z.open(filename) as f:
            js = f.read()
            df = pd.DataFrame(json.loads(js)["results"])
```
